```python
import requests
from bs4 import BeautifulSoup 
```

```python
url = "https://turbo.az/"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
cars = soup.find_all("div", {"class": "products-i__bottom"})
```

```python
for car in cars:
    print(car.find("div",{"class":"product-price"}).text)

    
```
